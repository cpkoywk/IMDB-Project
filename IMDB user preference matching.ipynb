# Movie preference matching I

I am tryna make a movie preferences matching system that matches two people's preferences' in movies, if it works well, it might be extended to match people's preferences in other media too like TV shows, songs, etc.

For example, if John and Jane both have seen some flicks, the number next to the movie title indicates their personal ratings for these films on a scale on 0-10

John:
Godfather 1: 9
Godfather 2: 10
Godfather 3: 6
Star Wars ep 1: 4
Star Wars ep 2: 3
Star Wars ep 3: 6
Mad Max Road Warrior: 8
Twilight: 2



Jane:
Godfather 1: 2
Godfather 2: 2
Godfather 3: 8
Star Wars ep 4: 8
Star Wars ep 5: 9
Star Wars ep 6: 5
Twilight: 10
Twilight 2: 8
Twilight 3: 9






```python
import pandas as pd
import numpy as np

john={'Godfather':9, 'Godfather 2':10,'Godfather 3': 6,'Star Wars ep 4': 4,'Star Wars ep 5': 3,'Star Wars ep 6': 6,
      'Mad Max Road Warrior': 8,'Twilight': 2}

jane={'Godfather':2, 'Godfather 2':2,'Godfather 3': 8, 'Star Wars ep 4': 8,'Star Wars ep 5': 9,'Star Wars ep 6': 5,
      'Twilight': 10,'Twilight 2': 8, 'Twilight 3': 9}

columns=['movies','score']

#create dataframes for these two people's movie ratings
johndf=pd.DataFrame(list(john.items()),columns=columns)

janedf=pd.DataFrame(list(jane.items()),columns=columns)

```

# Jaccard index for similarity matching

Jaccard Index takes both the number of movies seen and similarity into measure
basically

Jaccard Index = Intersect/Union

intersect of John and Jane: 7 (they have seen 7 movies in common)
union of John and Jane: 10 (John and Jane have seen 10 unique films)

so their Jaccard matching score is 0.7

```python
#Put it in code:
matchdf=pd.merge(johndf, janedf, on='movies')

matchdf2=pd.merge(johndf, janedf, on='movies',how='outer')

#intersect of John and Jane
intersect=matchdf.shape[0]

#union of John and Jane
union=matchdf2.shape[0]

#Jaccard score
jaccardscore=intersect/union

jaccardscore
```

# Adjusted Jaccard Score
but this doesn't take their drastic difference taste in the same movies into measure
it's unfair to say that they have the same preferences

so even though they have both seen godfather 1, it should not be considered the same for them.

John has rated Godfather 1 a 9, Jane Rated it a 2, so their similarity score:

Similarity score=1-|V1-V2|/10

1-7/10=0.3, godfather part one should only be considered 0.3 movie matching, instead of 1 movie matching.


```python
#Put it in code
matchdf['similarity']=1-abs(matchdf['score_x']-matchdf['score_y'])/10

intersect=sum(matchdf['similarity'])

print('of the {} movies John and Jane have both seen, their similarity score is {}.'.format(matchdf.shape[0],intersect))

#for union score, we count the uncommon movies as 1 (the ones only one of Jane or John has seen), 
#and the common ones as their similarity score

#the uncommon movies
matchdf3=matchdf2[matchdf2.isnull().any(axis=1)]

union=matchdf3.shape[0]+intersect

adjjaccardscore=intersect/union

print('Their adjusted jaccard score is {}.'.format(adjjaccardscore))
```

```python
# Lists to store the scraped data in
names = []
ids=[]
years = []
imdb_ratings = []
my_ratings = []
pages=[]

userid='ur21224949'
url = 'http://www.imdb.com/user/'+userid+'/ratings'
i=1
while True:
    try:
        pages.append(url)
        response = get(url)
        html_soup = BeautifulSoup(response.text, 'html.parser')
        newurl=html_soup.find('a',attrs={'class':"flat-button lister-page-next next-page"})
        movie_containers = html_soup.find_all('div', class_ = 'lister-item mode-detail')
        for container in movie_containers:
            # The name
            name = container.h3.a.text
            names.append(name)

            # The id
            id=container.find('div',attrs={'class':'lister-item-image ribbonize'})['data-tconst']
            ids.append(id)
            
            # When did you rate it?
            ratedate=container.find('p',attrs={'class':'text-muted'})

            # The year
            year = container.h3.find('span', class_ = 'lister-item-year').text
            years.append(year)

            # The IMDB rating
            imdb = float(container.find_all('span',attrs={'class':'ipl-rating-star__rating'})[0].text)
            imdb_ratings.append(imdb)
            
            # My Rating
            my_rating=float(container.find_all('span',attrs={'class':'ipl-rating-star__rating'})[1].text)
            my_ratings.append(my_rating)
        print ('page {} parsing done'.format(i))
        url='http://www.imdb.com'+newurl['href']
        i=i+1
        sleep(random.randint(1,3))
    except TypeError:
        print ('all pages parsing done, there are {} pages in total'.format(i))
        break
        
```
